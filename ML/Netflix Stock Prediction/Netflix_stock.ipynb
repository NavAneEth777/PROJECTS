```python
import pandas as pd
```

```python
df = pd.read_csv("NFLX.csv")
df.head()
```

```python
x = pd.to_datetime(df["Date"])
df["year"] = x.dt.year
df["month"] = x.dt.month
df["day"] = x.dt.day
df.head()
```

```python
df = df.drop("Date", axis="columns")
df.head()
```

```python
x = df.drop("Close", axis="columns")
x.head()
```

```python
y = df["Close"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=43)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn. metrics import r2_score

reg = LinearRegression()
reg.fit(x_train, y_train)
y_pred = reg.predict(x_test)
r2_score(y_pred, y_test)
```

```python
def prediction(year, month, date, Open, high, low, Adj_close, volume):
    features = [[year, month, date, Open, high, low, Adj_close, volume]]
    prediction = reg.predict(features).reshape(1,-1)
    
    return prediction[0]
```

```python
year = 2002
month = 5
date = 23
Open = 1.156429
high = 1.242857
low = 1.145714
Adj_close = 1.196429
volume = 104790000

predict = prediction(year, month, date, Open, high, low, Adj_close, volume)
print(predict)
```
